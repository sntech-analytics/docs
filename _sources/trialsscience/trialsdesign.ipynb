# Allocating samples in space and time

A trial is an experiment. Our aim is to contrast what would happen in the presence of or treatment to what would happen in the absence of the treatment. From a trials perspective, the challenge is to find the gear set up that will achieve this, and the unequivocally test the light vs non-light set up. A well-designed experiment must be unambiguous. When comparing treatment to control samples, the *only* thing that must differ between them is the treatment itself. 

If we get it wrong at the most fundamental stage, then we are wasting time, money, goodwill, and credibility.  

````{margin}
```{note}
It is tempting to consider the testing of effects such as Light/Non-light as a statistical issue. However the key is developing a setup that actually alters the environment you want to change in the way you intend it to, and to apply this change and compare to a *Control* state. This is about experimental design, not statistics. If an experimental design is critically flawed, no amount of statistics will save it.  
However, sometimes our experimental design will not be perfect. As long as the core is sound, we can use statistical methods to salvage weaker problems with the implementation of the protocol.
``` 
````
 

## Confounding. What it is, and how to avoid it.
````{margin}
```{note}
In its simplest form, all we are doing is comparing fish catches **with** *Pisces* to fish catches **without** *Pisces*. Herein lies the problem. We can't re-run time on exactly the same sample unit to see what the catch would have been with, and without the light. We have to compare the light treatment to a **Control**. But how do we ensure that the *only* thing that differed between treatment and control was the light itself? 
```
````
Consider a trial in which we have a single vessel, towing a single net. How do we test whether there are differences between Treatment and Control?  A naive approach might be to deploy Pisces at the start, do three tows with and three tows without Pisces, and see what the difference is.  

What is the problem with this? Let's think from the skipper's point of view. He heads out to the ground, deploys the trawl for 5 hours, lifts and see what he's got while deploying the next one. If the fishing is so-so, he probably won't decide to change location until about the third haul, and he may be steaming toward a different ground anyway.  
So... the skipper does three trawls at 3 knots for five hours each and in that time might have steamed 45 nautical miles from where he started. During that time we had the lights on (say). So, when we turn the lights off for the Control, what else has changed? We're in a different part of the ocean. Our test of light vs Control is *confounded* with Space. 

**Confounding** exists when the experimental manipulation cannot be separated from some other systematic different in the trial samples.  

````{margin}
```{warning}
Hey, if it was easy *everyone* would want to do it!
```
````
The most intuitive way to avoid this is to *intersperse* the treatments in some way. In the simple example above, we could simply alternate lights on and off. There is an additional issue to be aware of if doing this. Let's say the vessel does 4 tows a day, and we start each day with *light on*. What problem could this cause? In this case, the light treatment would *always* be on the pre-dawn and early afternoon tows, and the control would *always* be in the midday and evening tows. This introduces **temporal confounding**. Whenever there is a systematic element in the sampling program, we need to make sure we avoid any periodicity that will align with the sample program.  

One way of safeguarding against this is to change the starting treatment for each day's sampling program. Statistical purists might argue for a randomized start point, but as long as the starting points are interspersed across the trip, this is sufficient to avoid confounding. So using the example above, a 4 day trip could look like: 

```{list-table} Intersperse treatments in time and space to avoid confounding
:header-rows: 1
:name: treatment-table1

* - Day 1
  - Day 2
  - Day 3
  - Day 4
* - On
  - Off
  - On
  - Off
* - Off
  - On
  - Off
  - On
* - On
  - Off
  - On
  - Off
* - Off
  - On
  - Off
  - On  
```

For a single trawl, this is the only viable strategy to ensure interspersion. From a deployment point of view, this requires more work from the onboard observer/experimenter. Occasionally we might have to make allowances, but this spread of sampling should be the aim.  

As we add more light treatments into a trip, the sample design becomes more complex. The same principles apply, but it is preferable to have a defined plan rather than attempting to do it on the fly. For example, let's say we want to look at two colours. If we only had 4 days as above, what would the sample design look like?

```{list-table} Two-treatment levels
:header-rows: 1
:name: treatment-table2

* - Day 1
  - Day 2
  - Day 3
  - Day 4
* - Blue
  - Off
  - Green
  - Off
* - Off
  - Blue
  - Off
  - Green
* - Green
  - Off
  - Blue
  - Off
* - Off
  - Green
  - Off
  - Blue  

```
Do you see what we have done here? Each Colour treatment has been applied once in each position of the Day. This is somewhat remniscent of the old-school Latin-Square sample design dating back to Fisher (1926). However there is a twist. Each treatment is alternated with a Control. This will be used in the analysis as a blocking factor to deal with the short-range spatial confounding issue, as with the first design.  

```{note}
Interspersion of some kind is *always* a best practice in experimental design. Note however, that as we add more lights and combinations into the mix, an extra level of complexity is introduced. If we want to look at 3 colours with 3 flash combinations, for example, we might not be able to examine them all within a trip. This is where we make decisions about whether to squeeze the whole design into a single trip, or sequentially examine particular sets within a trip. Of course this introduces a higher level (between-trip) of statisical confounding. 
````

````{margin}
```{note}
The sample design is about addressing specific questions. There will always be some compromise and we can't let the demands of sample design 'elegance' outshadow what we are actually trying to achieve. A series of independent answers to the right questions is more important than an elegant but inefficient grand sampling plan, that is effectively unworkable.
```
````

## But wait... no two replicates are *exactly* the same!
Let's take a step back and think of what a *Control* is. The nomenclature is a bit unfortunate, and derives from the concept of a 'controlled experiment', in which all external factors are 'controlled' by the experimenter. This is the sort of thing we would do in engineering for example. We would keep all variables in a laboratory fixed, and manipulate only the one variable we were interested in to measure its effect.  

But ecological systems just don't work that way. This was the cue for a guy called Ronald Fisher who in  the 1920's realised that we don't need to physically control everything in the environment. We need to ensure that all the nuisance variables are random *with respect to the experimental treatment*. So it doesn't matter if we don't conduct all trial replicates on exactly the same tidal cycle, as long as replicates are sampled randomly across the range of possible tidal cycles.  

Now of course with low numbers of replicates in particular, a simple random allocation of treatments might just happen to align with periodicities such as tides. So we can force some interspersion to ensure that each combination of tides is sampled during each treatment across the experiment. This is the basis of the sample designs above. These extraneous factors are modeled in the analysis, and their effects estimated and removed and effectively corrected for.  

Recognising sources of confounding can be a challenge. But failure to do so invalidates the trial, and no amount of statistical magic can rescue it.

## Some general guidelines for single vs twin or multi-rigs

There are two features of a single trawl design that are, quite frankly, a pain in the ass.
1. Interspersion of the samples in time and space requires very careful planning
2. Even with effective interspersion, there will be a considerable element of spatio-temporal variability between treatment-control pairs. The centers of the trawls could be 15 knots apart if the vessel is steaming in a straight line.
3. The design takes a long time to implement. If we have only 16 trawls available over 4 days, then we have only 8 pairs of light-on/off as out level of replication.

```{note}
If at all possible, a twin-rig comparison should be used in preference to a single rig. 
````

If the vessel is a twin-rig outfit, life becomes a whole lot easier from the sample design point of view but, as we discuss later, it becomes more of an on-deck pain in the ass. However, the potential for confounding gains an extra level of complexity.  

Different sides of a vessel usually fish differently, but not necessarily consistently. For example, the Port net might be newer, with a slightly different height resulting in different catches. One of the vessels we work with has a ground in the Irish Sea over which he fishes in a 4 mile diameter circle, counter-clockwise. This introduces systematic differences, which may not be consistent in intensity depending on the tidal flow. Tricky, huh?

There are many reasons why different sides of a vessel may fish differently. The skipper will usually have a good sense of under which conditions different sides will perform, and it pays to listen to them. But keep an open mind.

The on-board conduct of a twin-rig trial requires some thought, planning, and some mental walk-through and rehearsals to work effectively. There are a few considerations:
- How many sets of lights, and how do we manage them?
Ideally in a twin-rig we have a set of lights on each and we turn them on and off as per the schedule.
- Can we turn one net off without inadvertently turning the other net off as well?
The lights use the same remote frequency. You need to consider how to ensure you don't screw up your treatments. At present, this problem is exacerbated by Helios - which currently uses the same remote control.
- If we only have one kit of Pisces available, how can we intersperse the treatments?
Moving the lights every single haul is a sure recipe for getting thrown off the boat into the North Sea. We need to develop an alternative strategy.
- If we only have one kit of Pisces, does this introduce confounding?
Yep. What differs between the treatment and control nets in addition to lighting? The physical presence of the lights themselves. While Pisces is not particularly heavy in water, it *can* distort the shape of the net particularly if deployed on the roof. Will it make a difference? Hopefully not.
- So, I have two nets of fish and/or *Nephrops* being dropped into the same hopper. Does it matter?
Yes... This is a tricky one to deal with and something that needs to be sorted out with the skipper and crew to minimise disruption. If SNT have hired the vessel, this can be dealt with at the hopper offerloading stage. If we are working in with the vessel's normal operation, the approach will depend on what you can sort out on the boat. On Virtuous, Tom has been known to jump into the hopper to get the correct subsample. On Golden Ray, we can generally sample the first haul early and wait for the catch to clear before getting the next sample.  
If we are fitting in with normal operations, the process will not be perfect so we just have to do the best we can.
