# Pisces placement on the gear

The trawl environment is a complex place. We are attempting to use light to alter the behaviour of fish at and within a device that is being towed at >3 knots in a dark environment that may or may not be obscured by dust clouds raised by the ground gear.  

Not all trawling gear is the same. The headrope of Nephrops-only trawls usually extend only 1-1.2m above the bottom. In contrast, a mixed fishery net may extend higher off the bottom. Bycatch reduction modifications are usually in place already - but are they effective? Are they visible to the fish or obscured by mud clouds?  

What follows is not an exposition of everything possible, but some examples of how we need to approach each problem with a fresh view to derive a workable strategy. In some ways, it's the most interesting and exciting part of the trial process.

```{figure} OtterTrawl.png
Anatomy of an otter trawl. A trawl is a complex physical structure, which generates its own microenvironment of water flow, light and sediment plumes. Understanding how it works is central to developing an effective Pisces trial.
```

```{attention}
The effectiveness of Pisces will be determined by the interaction between the gear, the light environment, and the fish's response to each of these.  
The key is to understand this interaction to determine the set-up that is most likely to induce an effect that can be exploited to reduce bycatch.
```

In general, there are two types of deployment to consider:
- Deployment at the net entrance to change catch composition
- Deployment within the net to separate the catch and facilitate escape  


## Deployment at the net entrance
Currently we deploy Pisces at the net entrance in an attempt to deter unwanted species of fish. Our trial results are in a very early stage, so we have no definitive answer about whether this will work. However, there are several considerations when suggesting this as a viable strategy

-  Can bycatch species respond in time, in a way that will enable the species to avoid the net?
Factors to consider are the towing speed of the net, the height above the bottom, the species, and the net construction. 
For example... Flatfish, skates, and rays rely on camouflage as protection. Regardless of their light responses in the laboratory, in the field the little buggers are not going to move until something is right on top of them.  In contrast, Gadoids such as cod and haddock have the ability to respond - but will they? Cod are renowned for being lazy and tracking with the net until they are exhausted, whereas haddock will move sideways and up when disturbed.

- Will the net construction allow for light to be visible in time for the fish to respond?
Two important factors to consider will be the headline height above the bottom, and the headline position relative to the ground gear.

Let's compare the standard Nephrops trawl with a coverless trawl. In both cases the height above the bottom is low. This means that any downward-facing lights will be bright near the bottom (which is good), but there will be little advance warning (which is bad!). What do we predict a fish to do if startled by the lights? A mobile fish *might* be able to avoid the headrope in a coverless trawl. But it will not receive as much warning as it would in a standard trawl. 

- Can we exploit the net construction to maximise the utility of Pisces?
There are quite a lot of legally permissible gear configurations out there in trawling-land. For various reasons (some reasonable, others not so) not all fishers adopt them. However when gear configuration allows there is potential to think outside the box during a trial.  

For example, one means of reducing bycatch of cod is by using a raised line. This configuration takes advantage of the behavioural difference between cod, which tend to stay near the bottom and haddock and whiting which tend to rise from the bottom on forst contact with fishing gear.

So now we should consider which light regimes we should try
There is no hard and fast rule here. But there are some factors to keep in mind:
- A trawl is moving quickly. The fish will not have much time to respond, and it will probably be a rapid reflex
- In general, fish like steady light unless it is blindingly bright
- A flash is more likely to startle a fish to get an immediate response
- Fish that live in dark environments probably perceive anything flash above 32Hz as a continuous light source
- Green, Blue and Cyan fall within the colour perception of most species we deal with on a trawl, so they are good bets
- Red is generally a waste of time in this fishery  
  
So this establishes a ball park of Pisces deployments to begin to look at. How can we tip the scales in our favour?  
One of the most important tools we have available to us in developing a strategy is Helios. Cameras deployed during the early protocol development phase are invaluable to observing and measuring responses of fish to the gear, and serve as a reality check that the Pisces deployment is visible to the fish. If we can't see Pisces illuminating the gear and the fish, the light stimulus for the fish will not be sufficient to induce any response.

In the Nephrops fishery the standard escape device is the Square Mesh Panel (SMP), which consists of mesh rotated at 45 degrees so it stays open during towing, and is usually located on the roof of the net just in front of the cod end.  

```{figure} SMP.png
:height: 350px
:name: SMP

The Square Mesh Panel is the standard escape device in the UK Nephrops trawl fishery
```

As a BRD, the SMP in its current configuration doesn't work as well as it should. On a Nephrops-only trawl with ~1.2m head height, on muddy bottom the SMP is quickly obscured by a dust cloud. In mixed gear on firmer bottom, the SMP is still visible however *if* the fish rises above the dust cloud. Most smaller fish tend to hug the bottom, so it is possible the panel is not effectively visible.  

The location of the SMP imediately in frontof the code end is also problematic. By the time the fish reaches the cod end it is tired, and even if it can see the SMP, the desire and energy level to squeeze through the panel are greatly reduced.  

However the SMP is not the only BRD available. In pelagic fleets in particular there are a range of other sorting devices such as Inclined Net Grids and Inclined Sorting Panels that are designed to separate catches either to move to different cod ends, or to escape gaps.

```{figure} InclinedNetGrid.png
:height: 350px
:name: InclinedNetGrid

Inclined Net Grid leading to escape gaps.
```

When developing a trial procedure, become familiar with the gear and think how to take advantage of light within the fishing environment.


## Summary
In this section I've raised some points about how and where we place Pisces to exploit the interaction between fish behaviour, gear confguration, and the light environment. This is not a prescriptive overview. Rather, it is intended to act as a sounding board when considering a trial and in particular how to deploy Pisces to be able to elicit a response.  

It is important to research the behaviour of the fishes, and understand the gear and how it operates. Cameras are invaluable, and their importance cannot be overstated. They will save time, money, and provide both the researcher and the skipper with confidence that Pisces deployment is not a waste of effort.
